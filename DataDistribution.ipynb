```python
import pickle as pkl
```

```python
X, y = pkl.load(open('ckpts/data_mnist_non_iid_1_num_workers_125_stratify_True_uniform_False_repeat_1.pkl', 'rb'))
```

```python
num_samples = []
worker_id = []
for i, (_) in enumerate(y):
    worker_id.append(i)
    num_samples.append(len(_))
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,3))
plt.bar(worker_id, sorted(num_samples), color='b')
plt.xlabel('devices')
plt.ylabel('num-samples')
plt.title('class-distribution')
plt.xlim(left=0, right=125)
plt.show()
```
